```python
import numpy as np
import sympy as sp
from sympy import integrate
from scipy.special import factorial



x = sp.symbols('x')
#__________________________modify_____________________________

a = -1
b = 1

def w(x):
    return 1 / sp.sqrt(1 - x**2)

n = 5
#_____________________________________________________________

a_coef = [0]
b_coef = [None, None]
P = []
P.append(1) # P_0(x) = 1

def tich_trong(f, g, a, b):
    return integrate(w(x)*f*g, (x, a, b))

def compute_a_coefficient(n):
    tmp = tich_trong(x * P[n - 1], P[n - 1], a, b) / tich_trong(P[n - 1], P[n - 1], a, b)
    a_coef.append(tmp)

def compute_b_coefficient(n):
    if n < 2:
        return
    tmp = tich_trong(x * P[n - 1], P[n - 2], a, b) / tich_trong(P[n - 2], P[n - 2], a, b)
    b_coef.append(tmp)

def P_n(x, n): # m = n - 1
    return (x - a_coef[n]) * P[n - 1] - b_coef[n] * P[n - 2]

def kiem_tra_truc_giao(Phi, n):
    mat = np.zeros((n + 1, n + 1))
    for i in range(0, n + 1):
        for j in range(0, n + 1):
            tmp = tich_trong(Phi[i], Phi[j], a, b)
            mat[i][j] = tmp
            print(tmp, end = "\t")
        print()


compute_a_coefficient(1)
P.append(x - a_coef[1]) # P_1(x) = x - a_1

for i in range(2, n + 1):
    compute_a_coefficient(i)
    compute_b_coefficient(i)
    P.append(P_n(x, i).simplify())

print("Các đa thức là:")
for i in range(n + 1):
    print(f'P_{i}(x) = {P[i]}')

print("Các hệ số a:")
for i in range(1, n + 1):
    print(f'a_{i} = {a_coef[i]}')

print("Các hệ số b:")
for i in range(2, n + 1):
    print(f'b_{i} = {b_coef[i]}')

print("Ma trận trực giao:")
kiem_tra_truc_giao(P, 5)
```
