```python
import sympy as sp 
from math import factorial
import numpy as np

x = sp.symbols('x')
c = sp.symbols('c')

def f(x):
    return 1 / (x + 1)

# get diffrential of f(x)
diff_f = [sp.diff(f(x), x, i) for i in range(1, 6)]

x_0 = 0

taylor = f(x_0) + diff_f[0].subs(x, x_0)*(x - x_0) + diff_f[1].subs(x, x_0)*(x - x_0)**2/2 + diff_f[2].subs(x, x_0)*(x - x_0)**3/6 + diff_f[3].subs(c, x_0)*(x - x_0)**4/24
```
