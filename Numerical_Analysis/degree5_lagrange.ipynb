# Degree 5 Lagrange function

```python
import matplotlib.pyplot as plt
import numpy as np
bar = np.linspace(-1, 1, 100)

L11 = 1/2 * (1 - bar)
L12 = 1/2 * (1 + bar)

L21 = 1/2 * bar * (bar - 1)
L22 = (1 - bar) * (1 + bar)
L23 = 1/2 * bar * (bar + 1)

L31 = -9/16 * ((bar + 1/3) * (bar - 1/3) * (bar - 1))
L32 = 27/16 * ((bar + 1) * (bar - 1/3) * (bar - 1))
L33 = -27/16 * ((bar + 1) * (bar + 1/3) * (bar - 1))
L34 = 9/16 * ((bar + 1) * (bar + 1/3) * (bar - 1/3))

L42 = -8/3 * ((bar + 1) * bar * (bar - 1/2) * (bar - 1))
L43 = 4 * ((bar + 1) * (bar + 1/2) * (bar - 1/2) * (bar - 1))
L44 = -8/3 * ((bar + 1) * (bar + 1/2) * bar * (bar - 1))
L45 = 2/3 * ((bar + 1) * (bar + 1/2) * bar * (bar - 1/2))

L51 = -625/768 * ((bar + 3/5) * (bar + 1/5) * (bar - 1/5) * (bar - 3/5) * (bar - 1))
L52 = 3125/768 * ((bar + 1) * (bar + 1/5) * (bar - 1/5) * (bar - 3/5) * (bar - 1))
L53 = -3125/384 * ((bar + 1) * (bar + 3/5) * (bar - 1/5) * (bar - 3/5) * (bar - 1))
L54 = 3125/384 * ((bar + 1) * (bar + 3/5) * (bar + 1/5) * (bar - 3/5) * (bar - 1))
L55 = -3125/768 * ((bar + 1) * (bar + 3/5) * (bar + 1/5) * (bar - 1/5) * (bar - 1))
L56 = 625/768 * ((bar + 1) * (bar + 3/5) * (bar + 1/5) * (bar - 1/5) * (bar - 3/5))

plt.figure(figsize=(16, 7))
plt.subplot(231)
plt.plot(bar, L11, label='L11')
plt.plot(bar, L12, label='L12')
plt.legend()
plt.axvline(x=0, color='k', linestyle='-')
plt.axhline(y=0, color='k', linestyle='-')
plt.grid()
plt.title('P1')

plt.subplot(232)
plt.plot(bar, L21, label='L21')
plt.plot(bar, L22, label='L22')
plt.plot(bar, L23, label='L23')
plt.legend()
plt.axvline(x=0, color='k', linestyle='-')
plt.axhline(y=0, color='k', linestyle='-')
plt.grid()
plt.title('P2')

plt.subplot(233)
plt.plot(bar, L31, label='L31')
plt.plot(bar, L32, label='L32')
plt.plot(bar, L33, label='L33')
plt.plot(bar, L34, label='L34')
plt.legend()
plt.axvline(x=0, color='k', linestyle='-')
plt.axhline(y=0, color='k', linestyle='-')
plt.grid()
plt.title('P3')

plt.subplot(234)
plt.plot(bar, L42, label='L42')
plt.plot(bar, L43, label='L43')
plt.plot(bar, L44, label='L44')
plt.plot(bar, L45, label='L45')
plt.legend()
plt.axvline(x=0, color='k', linestyle='-')
plt.axhline(y=0, color='k', linestyle='-')
plt.grid()
plt.title('P4')

plt.subplot(235)
plt.plot(bar, L51, label='L51')
plt.plot(bar, L52, label='L52')
plt.plot(bar, L53, label='L53')
plt.plot(bar, L54, label='L54')
plt.plot(bar, L55, label='L55')
plt.plot(bar, L56, label='L56')
plt.legend()
plt.axvline(x=0, color='k', linestyle='-')
plt.axhline(y=0, color='k', linestyle='-')
plt.grid()
plt.title('P5')

plt.tight_layout()
plt.show()
```
